```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot heatmap of missing values
plt.figure(figsize=(10, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
#vertical strips showing null values in the rows
```

```python
corr_matrix = df[['nsmiles', 'passengers', 'fare', 'large_ms', 'lf_ms']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Most popular routes by passenger volume
popular_routes = df.groupby(['city1', 'city2'])['passengers'].sum().sort_values(ascending=False).head(10)
print(popular_routes)
```

```python
# Compare fares and market share between the largest and lowest fare carriers
df[['carrier_lg', 'fare_lg', 'large_ms', 'carrier_low', 'fare_low', 'lf_ms']].groupby(['carrier_lg', 'carrier_low']).mean()
```

```python
#outlier Detection
sns.boxplot(data=df, y='fare')
plt.title('Outlier Detection in Fares')
plt.show()

sns.boxplot(data=df, y='passengers')
plt.title('Outlier Detection in Passenger Numbers')
plt.show()
```

```python
# Scatter plot between fare and distance
sns.scatterplot(data=df, x='nsmiles', y='fare')
plt.title('Fare vs. Distance')
plt.xlabel('Distance (Miles)')
plt.ylabel('Fare ($)')
plt.show()
```

Analyzing Seasonal Trends

```python
seasonal_trends = df.groupby(
    ['Year', 'quarter']
).agg(
    {
        'passengers': 'sum',
        'fare' : 'mean',
        'large_ms' : 'mean'
    }
)
print(seasonal_trends)
```

```python
seasonal_trends_1 = df.groupby('quarter').agg({'fare' : 'mean'})
print(seasonal_trends_1.round(2))
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=seasonal_trends, x='quarter', y='passengers', hue='Year', marker='o')
plt.title('Passenger Volume Trends by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Passengers')
plt.xticks(ticks=[1, 2, 3, 4], labels=['Q1', 'Q2', 'Q3', 'Q4'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=seasonal_trends, x='quarter', y='passengers', marker='o')
plt.title('Passenger Volume Trends by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Passengers')
plt.xticks(ticks=[1, 2, 3, 4], labels=['Q1', 'Q2', 'Q3', 'Q4'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=seasonal_trends, x='quarter', y='fare', marker='o')
plt.title('Average Fare Trends by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Average Fare')
plt.xticks(ticks=[1, 2, 3, 4], labels=['Q1', 'Q2', 'Q3', 'Q4'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=seasonal_trends, x='quarter', y='large_ms', marker='o')
plt.title('Market Share of Largest Carrier by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Market Share')
plt.xticks(ticks=[1, 2, 3, 4], labels=['Q1', 'Q2', 'Q3', 'Q4'])
plt.grid(True)
plt.show()
```
